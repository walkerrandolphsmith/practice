Now let's build our model!

```python
from sklearn.model_selection import train_test_split
seed_to_make_output_deterministic = 0
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=seed_to_make_output_deterministic)

print("Training data")
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

print("Test data")
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
import pandas as pd
import mglearn
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=1)

# train the model
classifier.fit(X_train, y_train)
```

```python
import numpy as np
# simulate new observations
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new shape: {}".format(X_new.shape))

# make prediction
prediction = classifier.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))
```

```python
# evaluate model accuracy with test data
y_pred = classifier.predict(X_test)
print("Test set actuals: {}".format(y_test))
print("Test set predictions: {}".format(y_pred))
where_predication_matches_actual_predicate = y_pred == y_test
print("Test set score: {:.2f}".format(np.mean(where_predication_matches_actual_predicate)))
print("Test set score with built in functionality: {:.2f}".format(classifier.score(X_test, y_test)))
```
